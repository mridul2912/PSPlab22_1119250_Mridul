```python
import pandapower as pp #import pandapower

net = pp.create_empty_network(sn_mva=100) #create an empty network


bus0 = pp.create_bus(net, name="MV Generator Internal Busbar", vn_kv=22, type="b")
bus1 = pp.create_bus(net, name="MV Generator Busbar", vn_kv=22, type="b")
bus2 = pp.create_bus(net, name="MV Load Busbar", vn_kv=11, type="b")
bus3 = pp.create_bus(net, name="HV Transformer1 Busbar", vn_kv=220, type="b")
bus4 = pp.create_bus(net, name="HV Transformer2 Busbar", vn_kv=220, type="b")
bus5 = pp.create_bus(net, name="HV Transformer3 Busbar", vn_kv=110, type="b")
bus6 = pp.create_bus(net, name="HV Transformer4 Busbar", vn_kv=110, type="b")

net.bus #shows the bus table

pp.create_ext_grid(net, bus0,vm_pu=1)

net.ext_grid

pp.create_load_from_cosphi(net, bus2, sn_mva=57, cos_phi=0.6, mode='underexcited', scaling=(11/10.45)**2, const_z_percent=100)

net.load

pp.create_impedance(net, bus0, bus1, rft_pu = 0, xft_pu = 0.18, sn_mva = 90)

net.impedance

trafo1 = pp.create_transformer_from_parameters(net, hv_bus = 3, lv_bus = 1, sn_mva = 50, vn_hv_kv = 220, vn_lv_kv = 22, vk_percent=10, vkr_percent=0, i0_percent=0, pfe_kw=0)
trafo2 = pp.create_transformer_from_parameters(net, hv_bus = 4, lv_bus = 2, sn_mva = 40, vn_hv_kv = 220, vn_lv_kv = 11, vk_percent=6, vkr_percent=0, i0_percent=0, pfe_kw=0)
trafo3 = pp.create_transformer_from_parameters(net, hv_bus = 5, lv_bus = 1, sn_mva = 40, vn_hv_kv = 110, vn_lv_kv = 22,vk_percent=6.4, vkr_percent=0, i0_percent=0, pfe_kw=0)
trafo4 = pp.create_transformer_from_parameters(net, hv_bus = 6, lv_bus = 2, sn_mva = 40, vn_hv_kv = 110, vn_lv_kv = 11,vk_percent=8, vkr_percent=0, i0_percent=0, pfe_kw=0)

net.trafo

line1 = pp.create_line_from_parameters(net, bus3, bus4, length_km = 1, r_ohm_per_km = 0, x_ohm_per_km = 48.4, c_nf_per_km = 0, max_i_ka=0.262)
line2 = pp.create_line_from_parameters(net, bus5, bus6, length_km = 1, r_ohm_per_km = 0, x_ohm_per_km = 65.43, c_nf_per_km = 0, max_i_ka=0.525)

net.line

pp.runpp(net, max_iteration = 100)

net.res_bus
net.res_trafo
net.res_line
net.res_load

```
